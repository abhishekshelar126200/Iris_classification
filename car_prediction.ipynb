```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("E:\python\CarPrice_Assignment.csv")
```

```python
df['fueltype1'] = np.where(df.iloc[:,3]=='gas',1,0)
df['aspiration1'] = np.where(df.iloc[:,4]=='std',1,0)
df['enginelocation'] = np.where(df.iloc[:,8]=='front',1,0)
```

```python
carbody_mapping={"convertible":1,"hatchback":2,"sedan":3,"wagon":4,"hardtop":5}
```

```python
df['carbody']=df['carbody'].replace(carbody_mapping)
```

```python
drivewheel_mapping={"rwd":1,"fwd":2,"4wd":3,}
```

```python
df['drivewheel']=df['drivewheel'].replace(drivewheel_mapping)
```

```python
enginetype_mapping={"dohc":1,"ohc":2,"l":3,"ohcv":4,"ohcf":5,"rotor":6,"dohcv":7}
```

```python
df['enginetype']=df['enginetype'].replace(enginetype_mapping)
```

```python
cylindernumber_mapping={"one":1,"two":2,"three":3,"four":4,"five":5,"six":6,"seven":7,"eight":8,"nine":9,"ten":10,"eleven":11,"twelve":12,}
```

```python
df['cylindernumber']=df['cylindernumber'].replace(cylindernumber_mapping)
```

```python
fuelsystem_mapping={"mpfi":1,"2bbl":2,"mfi":3,"1bbl":4,"spfi":5,"4bbl":6,"idi":7,"spdi":8}
```

```python
df['fuelsystem']=df['cylindernumber'].replace(fuelsystem_mapping)
```

```python
df.head()
```

```python
df.columns
```

```python
x=df[['symboling','fueltype1', 'aspiration1', 'carbody', 'drivewheel', 'enginelocation', 'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg',]]
```

```python
y=df['price']
```

```python
x.head(30)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
ytest.head(12)
```

```python
print(ypred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(ytest, ypred)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
mse = mean_squared_error(ytest, ypred)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
r_squared = r2_score(ytest, ypred)
print(f"R-squared (R2) coefficient: {r_squared}")
```
